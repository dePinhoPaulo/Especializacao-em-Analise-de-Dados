````python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)
````

````python
navegador.get(r"C:\Users\Kiko\Documents\Projetos Programação\Arquivos\Pagina Hashtag.html")
````

##### Ex1: Link do Whatsapp

````python
whatsapp = navegador.find_element(By.XPATH, '/html/body/footer/div/div[1]/div[2]/div/a[2]').get_attribute('href')
print(whatsapp)
````

##### Ex2: Imagens dos cursos

````python
imagem = navegador.find_element(By.XPATH, '/html/body/section[2]/div/div[4]/figure/a').get_attribute('href')
print(imagem)
````

#### Pegar informações de varios elementos da pagina

- links das imagens

````python
lista_imagens = navegador.find_elements(By.TAG_NAME, 'figure')
print(len(lista_imagens))
````

````python
lista_imagens = navegador.find_elements(By.TAG_NAME, 'figure')
for elemento in lista_imagens:   #Percorre as imagens da pagina
    link = elemento.find_element(By.TAG_NAME, 'a').get_attribute('href')
    print(link)
````

````python
lista_imagens = navegador.find_elements(By.TAG_NAME, 'figure')
for elemento in lista_imagens:   #Percorre as imagens da pagina
    links = elemento.find_elements(By.TAG_NAME, 'a')
    if links != []:   #pega somente as imagens que tem link
        link = elemento.find_element(By.TAG_NAME, 'a').get_attribute('href')    #retorna o link
        print(link)
````
