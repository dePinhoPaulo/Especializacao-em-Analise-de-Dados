### Dito isso, bora automatizar o envio de mensagens no Whatsapp

- Vamos usar o Selenium (vídeo da configuração na descrição)
- Temos 1 alternativa muito boa:
- Usar o https://web.whatsapp.com/send?phone=numero&text=texto (mais fácil, mais seguro, mas mais demorado)


```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time
import urllib
import os

servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)

navegador.get("https://web.whatsapp.com")

# esperar tela do whats carregar
# -> retorna lista com todos elementos com esse id
while len(navegador.find_elements(By.ID, 'side')) < 1 : # verifica se existe algun elemento com id da tela lateral, se não: espera 1 segundo
    time.sleep(1)
time.sleep(3) 
```

```python
# whatsapp ja carregou
import pandas as pd

tabela = pd.read_excel('Envios.xlsx')
tabela.info()
```

```python
# separando colunas em variaveis
for linha in tabela.index:
    nome = tabela.loc[linha, 'nome']
    mensagem = tabela.loc[linha, 'mensagem']
    arquivo = tabela.loc[linha, 'arquivo']
    telefone = tabela.loc[linha, 'telefone']

    print(f'nome: {nome}, "{texto}", n°: {telefone}')

    # tratando menssagem
    texto = mensagem.replace('fulano', nome)
    texto = urllib.parse.quote(texto)

    # enviando menssagem
    link = f'https://web.whatsapp.com/send?phone={telefone}&text={texto}'

    navegador.get(link)

    # espera whatsapp carregar -> tela lateral
    while len(navegador.find_elements(By.ID, 'side')) < 1 : # -> retorna lista com todos elementos com esse id
        time.sleep(1)  # verifica se existe algun elemento com id da tela lateral, se não: espera 1 segundo
    time.sleep(5) 

    # Se o numero não existir
    if len(navegador.find_elements(By.XPATH, '//*[@id="app"]/div/div/span[2]/div/span/div/div/div/div/div/div[1]')) < 1 :
        # envia menssagem
        navegador.find_element(By.XPATH, '//*[@id="main"]/footer/div[1]/div/span/div/div[2]/div/div[4]/div/span/div/div/div[1]').click()
        
        if arquivo != 'N':
            caminho_completo = os.path.abspath(f'arquivos/{arquivo}')
            time.sleep(3)
            navegador.find_element(By.XPATH, '//*[@id="main"]/footer/div[1]/div/span/div/div[2]/div/div[1]/div/span/div/div/div[1]/div[1]/span').click()
            time.sleep(3)
            navegador.find_element(By.XPATH, '//*[@id="app"]/div/div/span[6]/div/ul/div/div/div[1]/li/div/input').send_keys(caminho_completo)
            time.sleep(3)
            navegador.find_element(By.XPATH, '//*[@id="app"]/div/div/div[3]/div/div[3]/div[2]/div/span/div/div/div/div[2]/div/div[2]/div[2]/div/div').click()
            
        time.sleep(5)
```
