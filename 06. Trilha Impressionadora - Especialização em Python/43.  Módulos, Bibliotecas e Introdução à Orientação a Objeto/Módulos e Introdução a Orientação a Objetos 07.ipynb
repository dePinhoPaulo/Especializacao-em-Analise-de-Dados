# Mais edições de Gráfico com MatplotLib

```python
#importando o matplotlib
import matplotlib.pyplot as plt
```

- Usando o módulo numpy para trabalhar com números

```python
#importando o módulo numpy
import numpy as np
```

### Outros tipos de Gráfico:

- Linha

```python
vendas = np.random.randint(1000, 3000, 50)
meses = np.arange(1, 51)
print(vendas)
print(meses)
```

```python
plt.plot(meses, vendas)
plt.axis([1, 12, 0, max(vendas)+250])
plt.ylabel('vendas')
plt.xlabel('meses')
plt.plot()
```

- Editando o Gráfico de Linha

```python
plt.plot(meses, vendas, '-', color='red') #terceiro argumetno da função são os tipos de graficos (Documentação)
plt.axis([1, 50, 0, max(vendas)+250])
plt.ylabel('vendas')
plt.xlabel('meses')
plt.plot()
```

- Dispersão

```python
plt.scatter(meses, vendas)
plt.ylabel('vendas')
plt.xlabel('meses')
plt.plot()
```

- Barras

```python
plt.bar(meses, vendas)
plt.ylabel('vendas')
plt.xlabel('meses')
plt.plot()
```

- Trabalhando com Múltiplos Gráficos no mesmo "Plot" -> Para melhor visualização/comparação

```python
plt.figure(1, figsize=(15, 3))
plt.subplot(1, 3, 1)
plt.plot(meses, vendas, '-', color='red')
plt.subplot(1, 3, 2)
plt.scatter(meses, vendas)
plt.subplot(1, 3, 3)
plt.bar(meses, vendas)
```
