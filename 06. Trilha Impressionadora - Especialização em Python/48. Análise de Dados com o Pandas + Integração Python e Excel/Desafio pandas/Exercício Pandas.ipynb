### Importação de Módulos e Arquivos

```python
import pandas as pd

funcionario_df = pd.read_csv('CadastroFuncionarios.csv', sep=';', decimal=',')
cliente_df = pd.read_csv('CadastroClientes.csv', sep=';', decimal=',')
servico_df = pd.read_excel('BaseServiçosPrestados.xlsx')

#Retirar colunas estado civil e cargo da tabela funcionario
funcionario_df = funcionario_df.drop(['Estado Civil', 'Cargo'], axis=1) #Eixo 1 é colunas

display(funcionario_df)
display(cliente_df)
display(servico_df)
```

### 1 - Folha Salarial

```python
#Calcular a folha total
#forma 1:
salarios = funcionario_df[['Salario Base' , 'Impostos', 'Beneficios', 'VT', 'VR']].sum()
print(salarios)

valor_folha_total = salarios.sum()
print('Total da folha salarial é de R${:,}'.format(valor_folha_total))

#forma 2:
funcionario_df['Salario Total'] = funcionario_df[['Salario Base', 'Impostos', 'Beneficios', 'VT', 'VR']].sum(axis=1)
print('Total da folha salarial é de R${:,}'.format(funcionario_df['Salario Total'].sum()))
```

### 2 - Faturamento da Empresa

```python
faturamento_df = servico_df[['ID Cliente', 'Tempo Total de Contrato (Meses)']].merge(cliente_df[['ID Cliente', 'Valor Contrato Mensal']], on='ID Cliente')
display(faturamento_df)

faturamento_df['Faturamento Total'] = faturamento_df['Tempo Total de Contrato (Meses)'] * faturamento_df['Valor Contrato Mensal']
print('Faturamento Total R${:,}'.format(sum(faturamento_df['Faturamento Total'])))
```

### 3 - % Funcionários Fecharam Contrato

```python
qtd_funcionario_contratos = len(servico_df['ID Funcionário'].unique())
qtd_funcionario = len(funcionario_df['ID Funcionário'])

print('% Funcionario fecharam contratos: {:.2%}'.format(qtd_funcionario_contratos/qtd_funcionario))
```

### 4 - Qtde Contratos por Área

```python
contratos_area_df = servico_df[['ID Funcionário']].merge(funcionario_df[['ID Funcionário', 'Area']], on='ID Funcionário')
#display(contratos_area_df)

qtd_contratos_area = contratos_area_df['Area'].value_counts()
print(qtd_contratos_area)
```

### 5 - Funcionários por Área

```python
qtd_funcionarios_area = funcionario_df['Area'].value_counts()
print(qtd_funcionarios_area)

qtd_funcionarios_area.plot(kind='bar')
```

### 6 - Ticket Médio Mensal

```python
ticket_medio = cliente_df['Valor Contrato Mensal'].mean()

print('Ticket medio R${:,.2f}'.format(ticket_medio))
```
