# Temas do Seaborn

```python
import seaborn as sns

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

dados = {'Meses': meses, 'Vendas': vendas}

sns.lineplot(x='Meses', y='Vendas', data=dados)
```

```python
sns.set_theme(style='dark', context='notebook', palette='pastel')

sns.lineplot(x='Meses', y='Vendas', data=dados);
```

```python
import seaborn as sns

sns.set_theme(style='darkgrid', context='notebook', palette='colorblind')

# Dados fictcios
cidades = ['Braslia', 'Rio de Janeiro', 'So Paulo', 'Curitiba', 'Manaus']
temperaturas = [27, 29, 25, 23, 31]  # Temperaturas mdias anuais em graus Celsius

barras = sns.barplot(x=cidades, y=temperaturas, hue=cidades)
barras.set_title('Temperatura ao longo do ano')
barras.set_ylabel('Temperatura Celsius')
```
