## Gráfico de linha para apresentação


```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

plt.plot(meses, vendas, marker='o', linestyle='--', color='red', linewidth=2)
plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.show()
```

```python
#Gráfico apresentavel
import matplotlib.pyplot as plt
import numpy as np

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

marcadores = np.where(vendas == np.min(vendas), 'x', 'o') #retorna a lista de marcadores

plt.plot(meses, vendas)

for posicao, marcador in enumerate(marcadores): 
    if marcador == 'x':
        cor = 'red'
        tamanho = 80
    else:
        cor = 'blue'
        tamanho = 10
    plt.scatter(meses[posicao], vendas[posicao], marker=marcador, s=tamanho, color=cor)

plt.annotate('Menor Valor', color='red', xy=(2, 78), xytext=(2.4, 78))

plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.show()
```

```python
#Gráfico apresentavel pt2
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

marcadores = np.where(vendas == np.min(vendas), 'x', 'o') #retorna a lista de marcadores

cores = plt.get_cmap('Pastel2').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

plt.plot(meses, vendas)

for posicao, marcador in enumerate(marcadores): 
    if marcador == 'x':
        cor = 'red'
        tamanho = 80
    else:
        cor = 'black'
        tamanho = 0
    plt.scatter(meses[posicao], vendas[posicao], marker=marcador, s=tamanho, color=cor)
    plt.annotate(vendas[posicao], color=cor, xy=(meses[posicao], vendas[posicao]), xytext=(-5, -5), textcoords='offset points')

plt.annotate('Menor Valor', color='red', xy=(2, 78), xytext=(2.4, 78))

plt.axvspan(6, 11, alpha=0.15, color=cores[1])

plt.text(6, 80, 'Recuperação de Crescimento', color='green', fontsize=9, weight='bold')

plt.box(False)
plt.tick_params(axis='x', length=0)
plt.yticks([])

plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.show()
```
