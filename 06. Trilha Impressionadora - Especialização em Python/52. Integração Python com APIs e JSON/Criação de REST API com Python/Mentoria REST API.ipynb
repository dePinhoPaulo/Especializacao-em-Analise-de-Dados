### Criar uma API com Flask

```python
from flask import Flask
import pandas as pd

app = Flask(__name__) # Cria o site
tabela = pd.read_excel("Vendas - Dez.xlsx")

@app.route("/") # Decorator -> qual link a função vai rodar
def hello_world(): # Função
    return "<p>Hello, World!</p>"

@app.route("/polin") # Decorator -> qual link a função vai rodar
def polinho(): # Função
    return {'Nome': 'Paulo', 'Idade': 23, 'Time': 'Flamengo'} # retorna um JSON -> transformando em REST API

@app.route("/faturamento") 
def faturamento():
    faturamento = float(tabela['Valor Final'].sum())
    return {'Faturamento Total': faturamento} 

@app.route("/faturamento/produto") 
def faturamento_produto(): 
    vendas_produtos = tabela[['Produto', 'Valor Final']].groupby('Produto').sum()
    dic_vendas_prutos = vendas_produtos.to_dict()
    return dic_vendas_prutos 

@app.route("/faturamento/produto/<produto>") 
def fat_produto(produto): 
    tabela_vendas_produtos = tabela[['Produto', 'Valor Final']].groupby('Produto').sum()
    
    if produto in tabela_vendas_produtos.index:
        vendas_produto = tabela_vendas_produtos.loc[produto]
        dic_vendas_produto = vendas_produto.to_dict()
        return dic_vendas_produto
    else:
        return {produto: 'Inexistente'}

print(faturamento_produto())

app.run() # Roda o Site
```
